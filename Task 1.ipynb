```python
# Create a Bar Chart for Population Distribution

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create Dataset
data = {
    "Country Name": [
        "India",
        "China",
        "United States",
        "Indonesia",
        "Pakistan",
        "Nigeria",
        "Brazil",
        "Bangladesh",
        "Russia",
        "Mexico"
    ],
    "Population": [
        1450935791,
        1408280000,
        336000000,
        285000000,
        255000000,
        240000000,
        217000000,
        175000000,
        144000000,
        132000000
    ]
}

# Step 2: Create DataFrame
df = pd.DataFrame(data)

# Step 3: Save Dataset as CSV
df.to_csv("population_dataset.csv", index=False)

# Step 4: Display Dataset
print("Dataset Preview:")
print(df)

# Step 5: Create Bar Chart
plt.figure(figsize=(12,6))

plt.bar(
    df["Country Name"],
    df["Population"]
)

plt.title("Population Distribution by Country")
plt.xlabel("Country")
plt.ylabel("Population")

plt.xticks(rotation=45)

plt.tight_layout()

# Step 6: Save Chart
plt.savefig("population_distribution.png")

# Step 7: Show Chart
plt.show()

print("\nFiles Created Successfully:")
print("1. population_dataset.csv")
print("2. population_distribution.png")
```
